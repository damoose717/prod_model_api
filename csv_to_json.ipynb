```python
import csv 
import json 
import pandas as pd
import os
import subprocess
```

```python
def make_json(input_path, output_path): 
    """This function takes file paths as arguments and 
    converts CSV to JSON """      
    
    df = pd.read_csv(input_path)
    new_df = df.reset_index().to_json(orient='records')
    json_df = json.loads(new_df)
    
    for record in json_df:
        del record['index']
        
    with open(output_path, 'w') as f:
        json.dump(json_df, f, indent=4)
        
    with open(f'single_record_{output_path}', 'w') as f:
        json.dump(json_df[0], f, indent=4)
```

```python
input_path = 'exercise_26_test.csv'
output_path = 'exercise_26_test.json'
```

```python
if output_path in os.listdir():
    subprocess.run(['rm', '-r', output_path])

if f'single_record_{output_path}' in os.listdir():
    subprocess.run(['rm', '-r', f'single_record_{output_path}'])
    
make_json(input_path, output_path)
```
